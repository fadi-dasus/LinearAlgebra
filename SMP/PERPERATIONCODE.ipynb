```python
x = sp.Symbol('x')
fx = 1.5 * x **2
mean = sp.integrate(x*fx,(x,-1,1))
print('E(X) = ',mean)
var = sp.integrate(fx*x**2,(x,-1,1))-mean**2
var2 = sp.integrate(((x-mean)**2)*fx,(x,-1,1))-mean**2

print('Var(X) =' , var2,'or in the same way', var)
```
