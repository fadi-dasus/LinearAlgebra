```python
#Task 1
N=100
X = np.random.randn(N)*np.sqrt(2)+7 # as the variance is 2 the standard deviation is sqrt(2)
plt.plot(X,'*')
plt.show()
#Task 2
print("The sample mean is = " , X.mean())
# task 3
print("The sample variance is = " , np.sum((X-X.mean())**2)/(100-1))
# The mean is normally calculated as x.sum() / N, where N = len(x).
# If, however, ddof is specified, the divisor N - ddof is used instead.
# In standard statistical practice, ddof=1 provides an unbiased estimator of the variance of a
# hypothetical infinite population
print("The sample variance is = " ,X.var(ddof=1))
# Task 4
plt.figure()
plt.hist(X)
plt.show()
# Task 5
plt.figure()
stats.probplot(X, dist=stats.norm, plot=plt)
plt.show()
print("The points are mostly on a straight line meaning that the normal distribution is a good approximation.")
```

```python
#Task 1
prob = stats.norm.sf(4*60, loc=260, scale=50)
print(prob)
#Task 2
# ppf(q, loc=0, scale=1) Percent point function (inverse of cdf — percentiles)
quar25 = stats.norm.ppf(0.25, loc=260, scale=50)
quar75 = stats.norm.ppf(0.75, loc=260, scale=50)
print('quartiles 25%', quar25, 'minutes')
print('quartiles 75%', quar75, 'minutes')
```

```python
# Task 1
# this gives an array from 0 to 31
x=np.arange(0,32)

plt.plot(x,stats.binom.pmf(x, n=31, p=0.5),'*')
plt.show()
# Task 2
print("Yes the central limit theorem does apply since the sample size is greater than 31")
#Task 3

mean = stats.binom.mean(n=31, p=0.5) #31*0.5=15.5
std = stats.binom.std(n=31, p=0.5) # n*p*(1-p) = 31*0.5*(1-0.5) = 7.75
plt.plot(x,stats.norm.pdf(x, loc=mean, scale=std) ,'+')
plt.show()
print("The plots are almost identical meaning that normal distribution is a fine approximation. ")
```
