```python
import sympy as sp
```

```python
x = sp.Symbol('x')
y = sp.Symbol('y')
fxy = x + 3/2*y**2
fx = sp.integrate(fxy,(y,[0,1]))
print(fx)
print('to check if the result is true, we can integrate the f(x) for the limits from 0 to 1 and that should return 1 ')
check = sp.integrate(fx,(x,[0,1]))
print(check)

```

```python
x = sp.Symbol('x')
y = sp.Symbol('y')
fxy = x + 3/2*y**2
fx = sp.integrate(fxy,(y,[0,0.5]))
print(fx)
fy = sp.integrate(fx,(x,[0,0.5]))
print(fy)
```
