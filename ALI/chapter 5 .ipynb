```python
# to figure it out we need to check wether the system RREF(A- lI) is consistent and at least has a free variable
A = Matrix([[3,2],[3,8]])
x1 = sp.Symbol('x1')

l = 2
q = (A - 2*eye(A.shape[0])).rref()[0]
print('it is an eigen value for A because the system is consistent and it has a free variable')
q
```

```python
#  to find out we multiply V with A and the result should be a L[V]
x1 = sp.Symbol('x1')
x2 = sp.Symbol('x2')
A = Matrix([[1,-1],[6,-4]])
V = Matrix([[1],[3]])
print('check if AX=LX')
print('we need to find a value S.T A*V = ?*V if so we say that it is an eigenvector for A')
M = A*V
display(M)
print('when we compair the new vector with V we can see that the new victor is -2* V => the eignvalue is -2 ')
print('we can even verify by Rref the two victors and they must be linearly dependent like so')
print('we can even use the solve method to find the eigen value')
s = M.row_join(V).rref()[0]
display(s)
display(sp.solve_linear_system(s,x1,x2))

```

```python
l = sp.Symbol('lambda')
print('1) first we need to find lambda values ')

A = Matrix([[1,3],[4,2]])
lI =l*eye(A.shape[0])
n = A-lI
display(n)
# or d and eq gives the same equation
eq = det(A-lI)
sol = solve(eq,0,l)
# or
# eigenvalues = n.eigenvals()
display(sol)
print('so we have two distinct eigen value that means A is for sure diagonalized')
print('2) then we need to find the null space(eigen space) for all lambda values ')
print('for l = -2')
nulspace1 = (A --2*eye(A.shape[0])).nullspace()
nullsapace2 = (A -5*eye(A.shape[0])).nullspace()
display(nulspace1)
print('for l = 5')

display(nullsapace2)
print('3) we need to construct P and P-inverse')
P = nulspace1[0].row_join(nullsapace2[0])
display(P)
print('then P-inverse')
print('to be sure that P is inversible we check that detP != 0',det(P) != 0)
Pinv = P**-1
display(Pinv)
print('We need to construct D which is the lambda values on it diagonal')
D = Matrix([[-2,0],[0,5]])
display(D)
print('finally to test the result we should get A = P*D*Pinv')
P*D*Pinv == A
```
