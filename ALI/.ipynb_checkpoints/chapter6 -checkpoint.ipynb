```python
# 6.2 ex 9

u1 = Matrix([1,0,1])
u2 = Matrix([-1,4,1])
u3 = Matrix([2,1,-2])
x = Matrix([8,-4,-3])
# check if {u1,u2,u3} - orthogonal basis for R3
A = u1.row_join(u2).row_join(u3)
display(A.T*A)
print('since AT*A has only elements on its diagonal so it is an orthogonal basis')

# express 'x' as a linear combination of {u1,u2,u3}
# x = a1*u1 + a2*u2 + a3*u3


a1 = (x.dot(u1))/(u1.dot(u1))
a2 = (x.dot(u2))/(u2.dot(u2))
a3 = (x.dot(u3))/(u3.dot(u3))

display(a1*Symbol('u1') + a2*Symbol('u2') + a3*Symbol('u3'))
```

```python
# 6.3 ex 17

y = Matrix([4,8,1])
u1 = Matrix([Rational(2,3),Rational(1,3),Rational(2,3)])
u2 = Matrix([Rational(-2,3),Rational(2,3),Rational(1,3)])

# a. Let U = [u1 u2]. Compute UTU and UUT.
U = u1.row_join(u2)
print('since the result is I matrix then the matrix is an orthonormal basis and and U*U.T*Y is the projection vector')
# Since U.T*U is I2 => U is an orthonormal matrix => u1, u2 are orthonormal
display(U.T*U)
display(U*U.T)
display(y.project(u1) + y.project(u2))
display(U*U.T*y)
```
