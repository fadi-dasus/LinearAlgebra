```python
print('inverse the following matrix')
row1 = [1,2,1]
row2 = [4,5,3]
row3 = [0,0,2]
A = Matrix([row1,row2,row3])
display(A)
print('first we add the I to the matrix and rref() like so')
AI = A.row_join(sp.eye(A.shape[0])).rref()
display(AI)
# remove the I from AI after rref()
Ainv = AI[0][:,A.shape[0]:]
display(Ainv)
```

```python
A = Matrix([[1,-4,2,-1],[0,4,0,2],[3,-2,0,5],[0,1,0,7]])
display(A)
print('find the determinant, to figure out whether A is inversible and then find A**-1')
display(det(A))
print('this means that A has no free variables and the system is consistent and all columns are independent, and it spans R**4')
display(A.rref())
print('THE INVERSE OF A can be obtained by applying the following method {A-I RREF()= I-A**-1}')
I = Matrix(sp.eye(4))
Ainv = A.row_join(sp.eye(A.shape[0])).rref()[0][:,A.shape[0]:]
display(Ainv)
# or
# display(A.inv())
```

```python
A = Matrix([[3,0,4],[2,3,2],[0,5,-1]])
print('we can just use det(A)')
display(det(A))
# or
detA = A[0,1]*A.cofactor(0,1) + A[1,1]*A.cofactor(1,1) + A[2,1]*A.cofactor(2,1)
print('or we do this manually using the cofactor method')
display(detA)

B = Matrix([[0,5,1],[4,-3,0],[2,4,1]])
print('det B  = ')
display(det(B))
```

```python
# Find a such that det A = -18
a = sp.Symbol('a')
A = Matrix([
      [1, 5, -a]
    , [a, -a, a]
    , [2, 13, -7]
])
print('using the det method')
display(solve(A.det()+18,a))
# Find a such that det A = -18 using cofactor expansion a long the second row
A = Matrix([
      [1, 5, -a]
    , [a, -a, a]
    , [2, 13, -7]
])
print('using cofactor mathod')
solve(
A[1,0]*A.cofactor(1,0)+A[1,1]*A.cofactor(1,1)+A[1,2]*A.cofactor(1,2)+18,a
)
```

```python
# Find a such that det A = 24
a = sp.Symbol('a')
A = Matrix([
      [a, 3, -1, 0, -2]
    , [0, 2, -4, -1, -6]
    , [-2, -6, 2, 3, 9]
    , [3, 7, -3, 8, -7]
    , [3, 5, 5, 2, 7]  
])
solve(A.det()-24,a)
```
