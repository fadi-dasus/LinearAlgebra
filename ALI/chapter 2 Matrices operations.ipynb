```python
import sympy as sp
from sympy import *
from scipy import *
init_printing(use_latex = true)
from IPython.display import display, HTML,Math
import numpy as np
import pandas as pd

```

```python
print('inverse the following matrix')
row1 = [1,2,1]
row2 = [4,5,3]
row3 = [0,0,2]
A = Matrix([row1,row2,row3])
display(A)
print('first we add the I to the matrix like so')
I = Matrix(eye(3))
A_I = A.row_join(I).rref()
display(A_I)
```

```python
k = sp.Symbol('k')

A = Matrix([[2,3],[-1,1]])
B = Matrix([[1,9],[-3,k]])
AB = A*B
BA=B*A
print('AB - BA = 0 if they are the same ')
display(AB-BA)
print('3k + 6 = 0')
print('k + 2 = 0')
print('k = -2')
```
