```python
A = Matrix([[10,-8,-2,-2],[0,2,2,-2],[1,-1,6,0],[1,1,0,-2]])
W= Matrix([[2],[2],[0],[2]])
print('it is in the nul space because the multiplication of them is the 0 vector')
display(A*W)

print('if the vector W is in the column space it will nessesarly be a linear combination of the columns of A')
print('that means if we put them all in an augmented matrix and rref, then if we get a consistent system then is in the col space')
AW = A.row_join(W).rref()
display(AW)
print(' it is consistent then it is a linear combination of A vectors first column + the second columns = w')
W == A[:,0]+A[:,1]
```
